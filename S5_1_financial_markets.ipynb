# Financial markets
---

Financial markets are platforms where people, companies, and governments come together to trade financial assets—such as stocks (ownership shares in companies), bonds (loans to governments or corporations), and other instruments like currencies and commodities. Their main function is to channel funds from those who have them (savers/investors) to those who need them (businesses/governments), helping both sides meet their financial goals.

There are several types of financial markets, including:
- **Stock Markets**: Where company shares are bought and sold, enabling businesses to raise capital.
- **Bond Markets**: Where debt securities are issued and traded, allowing entities to borrow money at agreed-upon interest rates.
- **Foreign Exchange (Forex) Markets**: Where currencies are exchanged, vital for international trade and investment.
- **Commodity Markets**: Where raw materials (like oil and gold) are traded, playing a crucial role in global supply chains.

Key participants include retail investors, institutional investors (like pension funds and insurance companies), brokers/dealers (facilitating trades), and regulators (ensuring fair and transparent markets).

Financial markets are essential because they:
1. **Facilitate Capital Formation**: By matching investors looking for returns with organizations seeking funds for growth.  
2. **Enable Price Discovery**: Through supply and demand, markets help determine fair asset prices.  
3. **Enhance Liquidity**: Investors can quickly convert assets to cash or vice versa.  
4. **Distribute Risk**: Tools like derivatives let participants hedge or speculate on future price movements.

# Classification of Financial instruments
---

Financial instruments can be classified in various ways depending on their structure, maturity, and the nature of the rights they confer. Below is a broad classification:


### By **Nature of Claim** (What rights the instrument gives)

1. **Equity Instruments**  
   - Represent ownership in a company.  
   - Example: Common stocks (shares), preferred stocks.

2. **Debt Instruments**  
   - Represent a loan from the investor to the issuer (government, corporation).  
   - The issuer agrees to pay interest and repay principal at maturity.  
   - Example: Bonds, Treasury bills, commercial paper.

3. **Derivatives**  
   - Their value depends on (or “derives” from) an underlying asset (e.g., stocks, bonds, commodities, currencies).  
   - Used for hedging risk or speculating on price movements.  
   - Example: Futures, options, swaps, forwards.

4. **Hybrid Instruments**  
   - Combine features of both debt and equity or include embedded derivative features.  
   - Example: Convertible bonds (a bond that can be converted into a predetermined number of shares), preference shares with convertibility options.



### By **Maturity** (How long until repayment or expiration)

1. **Money Market Instruments**  
   - Short-term instruments (maturities of one year or less).  
   - Aim to provide liquidity and safety but usually offer lower returns.  
   - Example: Treasury bills (T-bills), commercial paper, certificates of deposit (CDs).

2. **Capital Market Instruments**  
   - Longer-term instruments (maturities greater than one year, or no maturity in the case of stocks).  
   - Aim for growth and/or income over a longer horizon.  
   - Example: Stocks, long-term bonds.



### By **Market of Issue/Trading**

1. **Primary Market**  
   - Where new securities are created and sold for the first time (e.g., an Initial Public Offering for a stock).  
   - Companies and governments raise fresh capital here.

2. **Secondary Market**  
   - Where existing securities are bought and sold among investors.  
   - Provides liquidity and price discovery for securities initially sold in the primary market.



### By **Underlying Asset** (For derivatives and certain specialized instruments)

1. **Equity-Based** (e.g., stock options, equity futures)  
2. **Debt-Based** (e.g., bond futures, interest rate swaps)  
3. **Commodity-Based** (e.g., gold, oil futures)  
4. **Foreign Exchange-Based** (e.g., currency futures, FX swaps)



### Why Classification Matters

- **Risk Management**: Different instruments carry different risk profiles (e.g., stock vs. bond risk).  
- **Investment Strategy**: Investors use a mix of short-term and long-term instruments to meet specific goals (liquidity, growth, income).  
- **Regulatory Oversight**: Authorities oversee instruments differently (e.g., stricter regulations for derivatives vs. standard bonds).  
- **Market Efficiency**: Clear classification helps in price discovery and transparency for market participants.



**In summary,** financial instruments can be categorized by the type of claim (equity, debt, derivative, or hybrid), by maturity (short-term vs. long-term), by market issuance (primary vs. secondary), or by underlying asset (for derivatives). This classification helps investors, issuers, and regulators understand and manage the risks, rewards, and purposes of each instrument.

# Overview of Intraday Trading

Intraday trading—often referred to as **day trading**—is the practice of buying and selling financial instruments (like stocks, currencies, or futures) within the same trading day. This means all positions are opened and closed before the market’s close to avoid overnight risk. Below is an overview:

<br>

### Key Characteristics

1. **Short Time Horizon**  
   - Trades typically last minutes or hours, closing before the trading session ends.
   
2. **Focus on Price Movements**  
   - Intraday traders rely on short-term price fluctuations rather than long-term fundamental trends.

3. **Frequent Trading**  
   - Multiple trades per day are common, requiring continuous market monitoring.

4. **Leverage and Margin**  
   - Many intraday traders use borrowed funds (margin) to amplify gains (and losses).

<br>

### Common Strategies

1. **Scalping**  
   - Multiple quick trades aimed at small, frequent profits.
   
2. **Momentum Trading**  
   - Targeting stocks moving strongly in one direction with high volume.

3. **Breakout Trading**  
   - Buying or selling when the price moves beyond identified support or resistance levels.

4. **Range Trading**  
   - Identifying price boundaries (support/resistance) and trading within the range.

5. **News/Events Trading**  
   - Taking advantage of price volatility immediately after announcements (earnings, economic data).

<br>

### Tools & Techniques

1. **Technical Analysis**  
   - Chart patterns, candlestick analysis, and technical indicators (e.g., Moving Averages, RSI) guide entry/exit decisions.

2. **Risk Management**  
   - Strict use of stop-loss orders to limit losses is vital because of the rapid price movements.

3. **Trading Platforms**  
   - Feature advanced charting tools, real-time quotes, and fast order executions.

4. **Discipline & Psychology**  
   - Emotional control and sticking to a well-defined trading plan are critical for consistency.

<br>

### Advantages

- **Immediate Results**: Trades settle quickly, and gains/losses are realized by day-end.  
- **No Overnight Risk**: Positions are closed before the market closes, avoiding news/events that might cause sharp price gaps.  
- **Potential for Quick Profits**: Rapid price moves can yield profits multiple times a day.

<br>

### Disadvantages

- **High Stress & Time-Intensive**: Requires constant market monitoring and fast decision-making.  
- **Increased Transaction Costs**: Frequent trades can incur significant commissions or spreads.  
- **High Risk**: Leverage amplifies losses as well as gains; volatile price swings can cause rapid drawdowns.  
- **Steep Learning Curve**: Success requires understanding technical tools, market psychology, and strict discipline.

<br>

#### Who Is It For?

Intraday trading appeals to individuals:
- Comfortable with **short-term risk**.  
- Interested in **actively monitoring** market movements.  
- With **time availability** during market hours.  
- Who can maintain **emotional discipline** under high-pressure conditions.

<br>

**In summary**, intraday trading involves rapid decision-making, technical analysis, and strict risk management. While it offers potential for quick gains and avoids overnight risk, its fast-paced nature and the high possibility of losses make it challenging for novices and demanding even for experienced traders.

# Financial Terminology
---



Below is a concise list of commonly used terminologies in financial markets, along with brief definitions. These terms can help you understand how trading and investing work.



1. **Ask Price**  
   - The lowest price at which a seller is willing to sell a security.

2. **Bid Price**  
   - The highest price a buyer is willing to pay for a security.

3. **Spread**  
   - The difference between the bid and ask price of a security, representing transaction cost or market liquidity.

4. **Bull Market**  
   - A market condition where prices are generally rising or expected to rise.

5. **Bear Market**  
   - A market condition where prices are generally falling or expected to fall.

6. **Liquidity**  
   - How easily an asset can be bought or sold in the market without significantly affecting its price.

7. **Volatility**  
   - The degree of variation in trading prices over time, indicating how quickly or dramatically prices move.

8. **Margin**  
   - Borrowed money used to purchase securities. Investors put up a percentage (margin) and borrow the rest from a broker.

9. **Leverage**  
   - The use of borrowed capital to increase the potential return on an investment. It also increases risk.

10. **Stock (Equity)**  
   - A share representing ownership in a company. Common stockholders often receive dividends and have voting rights.

11. **Bond (Debt Instrument)**  
   - A fixed-income instrument representing a loan made by an investor to a borrower (government, corporation). The borrower pays periodic interest and repays principal at maturity.

12. **Dividend**  
   - A portion of a company’s earnings distributed to shareholders (usually in cash, sometimes in additional shares).

13. **Yield**  
   - The income return on an investment (e.g., interest or dividends) expressed as a percentage of its price.

14. **Derivatives**  
   - Financial instruments whose value is derived from an underlying asset (e.g., stocks, bonds, commodities). Common examples are futures, options, and swaps.

15. **Option (Call/Put)**  
   - A contract giving the holder the right, but not the obligation, to buy (call) or sell (put) an underlying asset at a specified price within a specified time.

16. **Futures**  
   - A contract to buy or sell an asset (like oil or gold) at a future date and at a predetermined price, standardized and traded on exchanges.

17. **IPO (Initial Public Offering)**  
   - The process by which a private company first sells its shares to the public on a stock exchange to raise capital.

18. **Forex (Foreign Exchange)**  
   - The market where currencies are traded. It is the largest and most liquid market globally.

19. **ETF (Exchange-Traded Fund)**  
   - A basket of securities (stocks, bonds, or other assets) that trades on an exchange like a stock. Provides diversification and is often low-cost.

20. **Market Capitalization**  
   - The total market value of a company’s outstanding shares (calculated as share price × number of shares).

21. **Short Selling**  
   - Borrowing and selling a security with the expectation that its price will fall, allowing you to buy it back later at a lower price for a profit.

22. **Arbitrage**  
   - The practice of taking advantage of a price difference between two or more markets, profiting from the imbalance.

23. **Hedging**  
   - Implementing strategies (often using derivatives) to reduce or offset potential losses from price movements in an underlying asset.

24. **Order Types (Market/Limit)**  
   - **Market Order**: An order to buy or sell immediately at the best available price.  
   - **Limit Order**: An order to buy or sell at a specific price or better.

25. **Broker**  
   - A person or firm that acts as an intermediary, facilitating buy and sell orders for clients and often offering additional services like research.
