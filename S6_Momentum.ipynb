# Momentum Trading Strategies
---

A concise overview of momentum trading strategies — approaches to trading that leverage the tendency of rising assets to keep rising and falling assets to keep falling.



#### 1. What Is Momentum Trading?

Momentum trading involves buying securities that have shown strong upward price movement (positive momentum) or short-selling those demonstrating downward momentum. The central idea is that price trends, once in motion, can persist due to market psychology, institutional flows, and gradual information dissemination.

<br>

#### 2. Types of Momentum Strategies

1. **Time-Series (Trend-Following) Momentum**  
   - Looks at individual assets in isolation, comparing their current price to past prices.  
   - Example: If a stock’s price is above its 200-day moving average, go long; if it’s below, go short or remain out of the market.

2. **Cross-Sectional Momentum**  
   - Ranks a broad universe of assets (e.g., stocks) by their past returns over a defined period (e.g., 3, 6, or 12 months).  
   - Buys the top performers (the “winners”) and shorts the bottom performers (the “losers”) at each rebalance interval (e.g., monthly or quarterly).

<br>

#### 3. Common Momentum Indicators

1. **Moving Averages**  
   - Simple or Exponential Moving Averages help identify trend direction. Crossovers (shorter-term MA crossing above a longer-term MA) often signal the start of a new momentum phase.

2. **Relative Strength Index (RSI)**  
   - Although typically used to gauge overbought/oversold conditions, RSI can confirm momentum by remaining above (bullish) or below (bearish) a certain level (often 50).

3. **Rate of Change (ROC) / Momentum Oscillator**  
   - Measures the speed and magnitude of price changes. Positive, rising ROC indicates increasing upward momentum.

4. **Breakouts**  
   - Trading breakouts to new highs or lows can capture momentum surges, especially if accompanied by higher volume.

<br>

#### 4. Key Steps in a Momentum Strategy

1. **Define Lookback Period**  
   - Decide how far back you measure performance (e.g., 3-month, 6-month returns). Too short may lead to false signals; too long may miss rapid shifts.

2. **Establish Entry & Exit Rules**  
   - **Entry**: Enter once your momentum criteria are met (e.g., price crosses above a moving average, or top decile in cross-sectional ranking).  
   - **Exit**: Close the position if the price reverts below a certain threshold, or if the asset drops out of the top momentum tier.

3. **Position Sizing & Risk Management**  
   - Use stop-loss orders to limit downside if momentum fades.  
   - Diversify across multiple assets or sectors to mitigate the impact of sudden reversals.

4. **Rebalancing Frequency**  
   - For cross-sectional momentum, you might rebalance monthly or quarterly, recalculating momentum rankings and adjusting positions accordingly.

<br>

#### 5. Advantages

- **Systematic Rules**: Reduces emotional bias by following clear, data-driven signals.  
- **Adaptability**: Works across stocks, commodities, currencies, and bonds, allowing you to capture trends in various markets.  
- **Historical Evidence**: Academic studies (e.g., Jegadeesh & Titman) have shown momentum can generate excess returns over certain periods.

<br>

#### 6. Challenges and Risks

1. **High Volatility & Drawdowns**  
   - Sharp market reversals (often triggered by unexpected news or economic shifts) can lead to significant losses.  
   - Trend-following strategies can underperform in sideways or choppy markets.

2. **Transaction Costs**  
   - Frequent trading leads to potentially higher commissions and spreads.  
   - In cross-sectional momentum, rotating in and out of positions can be costly if not carefully managed.

3. **Crowded Trades**  
   - When many market participants chase the same momentum signals, trades can become crowded, reducing profitability and increasing slippage.

4. **Data & Overfitting**  
   - Historical backtests can be misleading if not validated on out-of-sample data or across different time frames.

<br>

#### 7. Best Practices

1. **Robust Testing**  
   - Use both in-sample (training) and out-of-sample (test) data to ensure your strategy isn’t just fitting noise.  
   - Conduct walk-forward analysis and Monte Carlo simulations if possible.

2. **Diversification**  
   - Applying the same momentum rules to different asset classes and time frames can help smooth returns.  
   - Avoid putting all capital into a single stock or sector trending strongly.

3. **Risk Controls**  
   - Combine momentum signals with volatility filters (e.g., exit or reduce position if volatility spikes).  
   - Use trailing stops to lock in gains while allowing the trend room to continue.

4. **Practical Constraints**  
   - Pay attention to liquidity: ensure the assets you trade can be entered/exited without large price impact.  
   - Factor in broker fees, short-selling restrictions, and potential regulatory constraints.

<br>

#### Key Takeaway

Momentum trading strategies aim to ride the wave of existing price trends, buying strength and selling weakness. While they can be lucrative, especially in trending markets, **risk management** and **cost control** are crucial given their susceptibility to sudden reversals and elevated trading frequency.

# Gap Strategy
---

**Gap up** and **gap down** are terms used in technical analysis to describe instances where a security’s opening price significantly differs from its previous closing price, creating a “gap” on the price chart. They often signal strong market sentiment changes or unexpected news/events. Here’s a closer look:

<br>

#### 1. Gap Up

- **Definition**: Occurs when a security opens at a price **higher** than its previous day’s closing price, leaving a noticeable gap on the chart.  
- **Possible Causes**:  
  1. **Positive News or Earnings**: Strong earnings reports or optimistic guidance can spark buying interest.  
  2. **Upgrades or Analyst Recommendations**: Favorable coverage from reputable analysts may boost investor sentiment.  
  3. **Overnight Developments**: Macroeconomic indicators (e.g., strong job data), geopolitical shifts, or after-market announcements.  
- **Implications for Traders**:  
  - A gap up **might** indicate bullish momentum if backed by high volume and sustained buying pressure.  
  - However, some traders watch for a possible **gap fill**—a pullback where the price returns to cover the gap area.

<br>

#### 2. Gap Down

- **Definition**: Occurs when a security opens at a price **lower** than its previous day’s closing price.  
- **Possible Causes**:  
  1. **Disappointing Earnings**: Missed earnings estimates, negative guidance, or lowered forecasts.  
  2. **Downgrades or Negative News**: Adverse analyst reports, product recalls, or lawsuits.  
  3. **Broader Market or Economic Factors**: Overnight market drops (e.g., global financial concerns) can cause widespread gap-down openings.  
- **Implications for Traders**:  
  - A gap down can **signal** bearish sentiment and potential further downside if supported by heavy selling volume.  
  - Some see gap downs as **buying opportunities** if the drop is viewed as an overreaction, anticipating a recovery to fill the gap.

<br>

#### 3. Types of Gaps

1. **Common Gap**  
   - Appears without any significant catalyst—often in quiet or low-volume periods.  
   - Typically filled relatively quickly as prices move back to previous levels.

2. **Breakaway Gap**  
   - Occurs at the start of a strong trend or after breaking a key support/resistance level.  
   - Often driven by meaningful news or a major chart pattern breakout.  
   - May not fill soon because it signifies a pronounced sentiment change.

3. **Runaway (Continuation) Gap**  
   - Appears during a strong, ongoing trend.  
   - Can indicate a healthy continuation of the move, often fueled by strong sentiment and momentum.

4. **Exhaustion Gap**  
   - Shows up late in a trend, potentially signaling a last surge of momentum before a reversal.  
   - May fill quickly if the market’s momentum fades or reverses soon after the gap.

<br>

#### 4. Trading Strategies Around Gaps

1. **Fade the Gap**  
   - Traders assume the initial gap will reverse (fill) quickly.  
   - They short a gap-up or go long on a gap-down, expecting a return to prior day’s close.

2. **Trade the Break**  
   - Momentum traders buy a gap-up (or short a gap-down) if it’s supported by high volume, continuing the trend.

3. **Gap-and-Go**  
   - Looks for a fast continuation right after the opening gap if the underlying cause is strong (positive earnings, bullish market sentiment).

4. **Gap Reversal Patterns**  
   - Wait for confirmation (like a reversal candlestick or a significant volume shift) before trading against the gap direction.

<br>

#### Key Takeaways

- **Gap ups** and **gap downs** often signal shifts in sentiment or new information entering the market.  
- Volume, overall market context, and follow-through price action help determine if a gap indicates a **trend continuation**, a **reversal**, or just a short-term **overreaction**.  
- Understanding **why** the gap occurred (earnings news, analyst upgrade/downgrade, broader market events) is crucial in deciding whether to **fade** the gap or **trade** it in the direction of momentum.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf

import warnings
warnings.filterwarnings("ignore")
```

<br>

First, lets grab some data

```python
start  = '2018-09-04'
end    = '2019-08-30'

df = yf.download("AAPL", start=start, end=end, auto_adjust=False)
df.head()
```

We need to calculate the adjusted open price. The adjusted open price is determined by multiplying the open price by an adjustment factor,  calculated by comparing the adjusted close to the actual close price.

```python
df = df.stack(level=1).rename_axis(['Date', 'Ticker']).reset_index(level=1)
```

```python
adj_fac = df['Adj Close'] / df['Close']
```

We can now use this to calcultae the adjusted open price and store in our dataframe

```python
df['Adj Open'] = adj_fac * df['Open']
df.head(3)
```

### Take a Position

We will initiate **long positions** if today’s opening price exceeds yesterday’s closing price (gap up), and **short positions** if today’s opening price falls below yesterday’s closing price (gap down).

**Long Position**: Today’s open > Yesterday’s close  
**Short Position**: Today’s open < Yesterday’s close

Let's create a new column to store the position we will take

```python
df['position'] = np.nan
```

```python
df.head()
```

Store your long positions

```python
long = df['Adj Open'] > df['Adj Close'].shift(1)
```

Store your short positions

```python
short = df['Adj Open'] < df['Adj Close'].shift(1)
```

Set the position labels to the following and clean up the nan values

```python
df.loc[long,  'position'] = 1
df.loc[short, 'position'] = -1

df = df.dropna()
```

```python
df.head()
```

Calculate the returns

```python
returns = (df['Adj Close'] - df['Adj Open']) / df['Adj Open']
```

Now, calculate the strategy returns

```python
df['strat_returns'] = returns * df.position
```

...and the cumulative strategy returns

```python
cum_strat_returns = (df.strat_returns + 1).cumprod()
```

We will also shows the returns of buying and holding over the same period

```python
daily_returns = df['Adj Close'].pct_change()
cum_returns   = (daily_returns + 1).cumprod()
```

Now, let's plot the results

```python
plt.figure(figsize=(16,8))
plt.plot(cum_strat_returns, label='Momentum')
plt.plot(cum_returns, label='Buy & Hold')
plt.grid()

plt.legend()
plt.title('Cumulative strategy returns of Gap Strategy v Buy and hold', fontsize=18)
plt.xlabel('Date')
plt.ylabel('Cumulative returns');
```

As you can see the momentum strategy performed better than just buying and holding (excluding costs etc).
